```python
X=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
X
```

```python
y=df['Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split #from sklearn.family import model
```

```python
train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=101)#tuple unpacking to gain training and testing set
#training set given 60% and testing set given 40%
#random_state given just because to match the samples for other person if uses the same number his sample space will be same
```

```python
from sklearn.linear_model import LinearRegression #from sklearn.family import model
```

```python
lm = LinearRegression()  #invoked the LinearRegression object
```

```python
lm.fit(X_train,y_train)  #used to fit the training set data's and creating hypothesis for training the data
```

```python
lm.intercept_
```

```python
lm.coef_#this is theta array of the linear regression model for relating 
#suppose our example has only one parameter we can use hypothesis function h=theta(0)+theta(1)X and h-y 
#will  be the error
```

```python
cdf = pd.DataFrame(lm.coef_,X_train.columns,columns=['Coefficient'])
```

```python
cdf # for 1 unit increase in any one of the index value will lead to an increase of the coefficient with the value of the coefficient
```

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(y_test,predictions,markers=['O','X'])
```

```python
plt.figure(figsize=(12,7))             
sns.distplot((y_test-predictions),color='red')   #uses the y_testing and predictions to give whether our predictions are having greater recall as well as higher precision
#here the error between y_test and predictions plots normal distribution hence the process is accurate
```

```python
sns.set_context('notebook')
sns.jointplot(data=customers,x='Time on Website',y='Yearly Amount Spent')
```

```python
customers
```

```python
X = customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership' ]]
```

```python
y=customers['Yearly Amount Spent']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
lm.predict(X_test)
```

```python
sns.scatterplot(x=lm.predict(X_test),y=y_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,lm.predict(X_test))
```

```python
metrics.mean_squared_error(y_test,lm.predict(X_test))
```

```python
np.sqrt(metrics.mean_squared_error(y_test,lm.predict(X_test)))
```

```python
lm.score(X_test,y_test)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(a=y_test-(lm.predict(X_test)),color='red')
```
