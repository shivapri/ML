```python
def impute_ags(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            return  np.round(38.233441)
        elif Pclass==2:
            return np.round(29.877630)
        elif Pclass==3:
            return np.round(25.140620)
    else:
        return Age
            
```

```python
np.round(29.877630)
```

```python
train['Age']=train[['Age','Pclass']].apply(impute_ags,axis=1)
```

```python
sns.heatmap(data=train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
